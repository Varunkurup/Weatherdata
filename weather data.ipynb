```python
def mean_for_mth(feature):
    mean = []
    for x in range(12):
        mean.append(
            float("{:.2f}".format(weather_data[weather_data['MONTH'] == (x+1)][feature].mean())))
    return mean
```

```python
labels_basel = labels['BASEL_BBQ_weather']
labels_basel.head()
```

```python
true_labels = len(labels_basel[labels_basel==False])
false_labels = len(labels_basel[labels_basel==True])
print(true_labels)
print(false_labels)
print('Precent of True values: {0:.2f}%'.format(true_labels/(true_labels+false_labels)*100))
print('Precent of False values: {0:.2f}%'.format(false_labels/(false_labels+true_labels)*100))
```

```python
labels_basel = labels_basel.astype(int)
from imblearn.over_sampling import RandomOverSampler
oversample = RandomOverSampler()
ovrspl_X, ovrspl_y  = oversample.fit_resample(basel_x, labels_basel)
```

```python
from sklearn import preprocessing
x_train = basel_x
x_train = x_train =preprocessing.StandardScaler().fit(x_train).transform(x_train)
y_train = labels_basel
```

```python
from sklearn.model_selection import train_test_split
x_train1,x_test1,y_train1,y_test1 = train_test_split(x_train,y_train,test_size =  0.3,random_state=1,stratify=y_train)
```

```python
from sklearn import svm
from sklearn import metrics
svs=['rbf','linear','poly','sigmoid']
svm_test_accuracy=[]
for s in svs:
    clf=svm.SVC(kernel=s)
    clf.fit(x_train1,y_train1)
    svm_test_accuracy.append(metrics.accuracy_score(y_test1,clf.predict(x_test1)))
svm_test_accuracy
```
